## Task 3

train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
l_r=LinearRegression()
```

```python
l_r.fit(x_train,y_train)
```

```python
y_pred=l_r.predict(x_test)
y_pred
```

```python
#Evaluation
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error,r2_score
```

```python
print("mean_absolute_error:--")
l_m_a_e=mean_absolute_error(y_pred,y_test)
print(l_m_a_e)
```

```python
print("root_mean_squared_error")
lr_m_s_e=root_mean_squared_error(y_pred,y_test)
print(lr_m_s_e)
```

```python
print("r2_score:--")
l_r2_score=r2_score(y_pred,y_test)
print(l_r2_score)
```

Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings("ignore")
```

```python
r_f=RandomForestRegressor()
```

```python
r_f.fit(x_train,y_train)
```

```python
y_predd=r_f.predict(x_test)
y_predd
```

```python
print("mean_absolute_error:--")
r_m_a_e=mean_absolute_error(y_predd,y_test)
print(r_m_a_e)
```

```python
print("root_mean_squared_error")
r_m_s_e=root_mean_squared_error(y_predd,y_test)
print(r_m_s_e)
```

```python
print("r2_score:--")
r_r2_score=r2_score(y_predd,y_test)
print(r_r2_score)
```

```python
comparison = pd.DataFrame({
    'Model': ['Random Forest','Linear Regression'],
    'MAE': [r_m_a_e,l_m_a_e],
    'RMSE': [r_m_s_e, lr_m_s_e],
    'R2 Score': [r_r2_score, l_r2_score]
})

print(comparison)
```

## Insights & Summary

The analysis showed that features such as area, number of bathrooms, air conditioning, and preferred location had the strongest influence on house prices. Among the two models tested, Linear Regression performed slightly better than Random Forest, achieving a lower MAE (970,043), lower RMSE (1,324,507), and a higher R² score (0.273). This means the model was able to explain about 27% of the variation in house prices, indicating moderate predictive performance. One surprising finding was that amenities such as air conditioning and location preferences significantly affected house prices in addition to the size of the property. The distribution of house prices was also somewhat skewed, with a few high-priced properties influencing the overall range. Based on these findings, real estate businesses should focus on promoting properties with larger areas, modern amenities, and desirable locations, as these factors have the greatest impact on market value.
